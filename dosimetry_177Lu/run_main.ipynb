**In this file one the results for the dosimetry code are presented.**

The main function from the *main.py* file needs the activity dictionary (with real and simulated activities per time point), the *.nrrd/.nii* files for the 24h calibrated (and registered) volumes, as well as the segmentation. 

The user needs to define the injected activity (MBq), to get the normalized mean doses. The generated absorbed dose maps are normalized to the administered activity. To override this, 
one can set the administered activity to 1 MBq. The DVHs are, however, generated in mGy.

An output path needs to be set to save the generated files.

The *Time-Integrated Activity* (TIA) maps, the absorbed dose maps (mGy/MBq) and a .txt file with the statistics for absorbed dose are generated for both real and simulated distributions and saved in the folder provided as the output path.
